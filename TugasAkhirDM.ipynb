```python
# Import Library yang digunakan 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Menyiapkan data dan memanggil dataset
dataset = pd.read_csv('ProduksiPadi_NTB.csv')
dataset.keys()
```

```python
# Menampilkan 5 baris data pertama dataset
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
# Mengaktifkan K-Means dengan jumlah K=2
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
# Menampilkan nilai Centroid yang digenerate oleh algoritma
print(kmeans.cluster_centers_)
```

```python
# Plot Data Point
# Memvisualisasikan Hasil Klasterisasi Data Konsumen
plt.scatter(X[:,0], X[:,11], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Tahun")
plt.ylabel("Rata-rata produksi")
plt.title("Grafik Hasil Klasterisasi Rata-rata produksi per tahun")
plt.show()
```

```python
# Plot Data Point
# Memvisualisasikan Hasil Klasterisasi dengan centroid dari masing masing klaster
plt.scatter(X[:,0], X[:,11], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.xlabel("Tahun")
plt.ylabel("Rata-rata produksi")
plt.title("Grafik Hasil Klasterisasi Data Rata-rata produksi per tahun dengan Centroid")
plt.show()
```
